## Number of CPU for Socket Type

```python
fig = plt.figure(figsize=(100,10))
df1['socket'].value_counts().plot.bar(color=['red', 'blue', 'green', 'yellow', 'orange', 'purple'])
plt.title('Number of CPUs per Socket type')
plt.show()
```

```python
# Too many socket types to see, so we will only view the top ones 
top30sockets = df1.socket.value_counts()
```

```python
top30sockets
```

```python
top30sockets_test = df1['socket'].value_counts()
```

```python
topSockets = df1[~df1['socket'].isin(top30sockets_test[top30sockets_test < 10].index)]
topSockets
```

```python
fig = plt.figure(figsize=(20,10))
topSockets['socket'].value_counts().plot.bar(color=['red', 'blue', 'green', 'yellow', 'orange'])
plt.title('Number of CPUs per Socket type')
plt.show()
```

#### We will now remove the unknowns from the figure above 

```python
topSockets_clean = topSockets[topSockets.socket != 'unknown']
topSockets_clean
```

```python
fig = plt.figure(figsize=(20,10))
topSockets_clean['socket'].value_counts().plot.bar(color=['red', 'blue', 'green', 'yellow', 'orange'])
plt.title('Number of CPUs per Socket type, cleaned')
plt.show()
```

## Socket Types 
#### Using the figure above we can see that AM4 seems to be the most popular socket type used.
